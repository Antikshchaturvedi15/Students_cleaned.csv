```python
import pandas as pd
```

```python
df = pd.read_csv("./Data/Raw/Processed/student.csv")
```

```python
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")
```

```python
df.loc[df["Age"] > 50,"Age"] = pd.NA
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Age"] = df["Age"].astype(int)
```

```python
df.loc[df["Age"] < 0, "Age"] = pd.NA
```

```python
df["Gender"] = df["Gender"].replace({
    "M": "Male",
    "F": "Female",
    "Unknown": None
})
```

```python
df["Email"] = df["Email"].replace({"NaN":"None"})
```

```python
df["Course"] = df["Course"].replace("", pd.NA)
df["Email"] = df["Email"].replace("", pd.NA)

df["Course"] = df["Course"].fillna("Not Assigned")
df["Email"] = df["Email"].fillna("No Email")
```

```python
df["Course"] = df["Course"].replace("", pd.NA)
df["Course"] = df["Course"].fillna("Not Assigned")
```
